```python
import cv2
import numpy as np
from scipy.signal import fftconvolve
from scipy.signal import gaussian
import os

input_folder = "/home/muhammad/Downloads/umairbhaiwork/Boat Detection/boat img data/ship"
output_folder = "/home/muhammad/Downloads/umairbhaiwork/Boat Detection/boat img data/ship/deblurred_images"
target_size = (256, 256)
scale_up_threshold = 512

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

def scale_image(image_path, target_size=(256, 256), scale_up_threshold=512):
    img = cv2.imread(image_path)
    height, width, _ = img.shape

    if width < scale_up_threshold or height < scale_up_threshold:
        # Scale up the image
        scale_factor = max(target_size[0] / width, target_size[1] / height)
        new_width = int(width * scale_factor)
        new_height = int(height * scale_factor)
        img = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_LANCZOS4)

    img = cv2.resize(img, target_size, interpolation=cv2.INTER_LANCZOS4)
    return img



for filename in os.listdir(input_folder):
    if filename.lower().endswith(('bmp','.png', '.jpg', '.jpeg')):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)
        
        img = cv2.imread(input_path)
        
        if img.shape[0] < target_size[0] or img.shape[1] < target_size[1]:
            scaled_img = scale_image(input_path, target_size, scale_up_threshold)
            
            cv2.imwrite(output_path, scaled_img)
        else:
            cv2.imwrite(output_path, img)

print("Images deblurred and saved to", output_folder)
```
