# Model Predicton

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = load_model(os.path.join('models','shipimageclassifierlive.h5'))
```

## Ship Detection

```python
img = cv2.imread('testimg/tb4.jpeg')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
resize = tf.image.resize(img, (256, 256))

plt.imshow(resize.numpy().astype(int))

plt.show()
```

```python
yhat_newmodel = model.predict(np.expand_dims(resize/255, 0))
```

```python
yhat_newmodel
```

```python
if yhat_newmodel >0.5:
    print(f'Predicted class is ship')
else:
    print(f'Predicted class is not ship')
```
